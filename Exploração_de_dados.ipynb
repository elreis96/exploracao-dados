# Carregamento dos Dados


```python
# Carregamento dos Dados

df_customers = pd.read_csv('/content/customers_data.csv')
df_carbon = pd.read_csv('/content/carbon_offsets.csv')
df_rentals = pd.read_json('/content/rentals_data.json')
df_vehicles = pd.read_json('/content/vehicles_data.json')
```

# Identificação de dados






```python
df_customers.columns
df_customers.info()
df_customers.head()
display(df_customers)

df_carbon.columns
df_carbon.info()
df_carbon.head()
display(df_carbon)

df_rentals.columns
df_rentals.info()
df_rentals.head()
display(df_rentals)

df_vehicles.columns
df_vehicles.info()
df_vehicles.head()
display(df_vehicles)
```

# Correções


1. Customers - age - valor ausente
2. Rentals - daily_rate com valor < 0
3. Rentals - daily_rate como string
4. Rentals - daily_rate como negativo
5. Rentals - data invalida
5. Rentals - data formatação errada
6. Vehicles - co2_g_per_km dado ausente


```python
# Customers - age - valor ausente
media_idade = df_customers['age'].mean()

df_customers['age'] = df_customers['age'].fillna(media_idade)


display(df_customers)
```

```python
# Rentals - daily_rate como string
df_rentals['daily_rate'] = pd.to_numeric(df_rentals['daily_rate'], errors='coerce')
#media_daily_rate = df_rentals['daily_rate'].mean()
#df_rentals['daily_rate'] = df_rentals['daily_rate'].fillna(media_daily_rate)

# Rentals - daily_rate com valor < 0
df_rentals.loc[df_rentals[ 'daily_rate'] <= 0, 'daily_rate'] = pd.NA

dados_truncados = df_rentals[['start_date', 'end_date', 'daily_rate']].isna().sum()
print("Resumo de dados ausentes/inválidos:")
print(dados_truncados)


df_rentals['daily_rate'] = df_rentals['daily_rate'].replace("", pd.NA)

# Rentals - data invalida
df_rentals['start_date'] = pd.to_datetime(df_rentals['start_date'], errors='coerce')
df_rentals['end_date'] = pd.to_datetime(df_rentals['end_date'], errors='coerce')

# Rentals - data formatação errada


display(df_rentals)

```

```python
# Vehicles - co2_g_per_km dado ausente
df_vehicles['co2_g_per_km'] = df_vehicles['co2_g_per_km'].fillna(120)
```

# Engenharia de Dados


- Calcular a duração de cada locação(em dias).
- Calcular a receita total por locação.
- Merge com a base de veículos.
- Cálculo da emissão de CO₂ por locação.
- Extração de ano e mês da data de início da locação.

```python
# Criaçao do Backup da base de dados
df_carbon_clean = df_carbon
df_rentals_clean = df_rentals
df_customers_clean = df_customers
df_vehicles_clean = df_vehicles
```

```python
df_rentals_clean['dias_locacao'] = (
    df_rentals_clean['end_date'] - df_rentals_clean['start_date']
    ).dt.days
display(df_rentals_clean)
```

```python
# Calculo da receita total por locação
# Valor da diária = daily_rate
# Dias de locação * valor da diária

df_rentals_clean['receita_total'] = df_rentals_clean['dias_locacao']*df_rentals_clean['daily_rate']
```

```python
# Merge com a basa de veículos

df_consolidado = df_rentals_clean.merge(df_vehicles_clean, on ='vehicle_id')

df_consolidado = df_consolidado.merge(df_customers_clean, on = 'customer_id')
```

```python
# Cálculo da emissão de CO₂ por locação
df_consolidado['co2_emitido'] = df_consolidado['co2_g_per_km'] * df_consolidado['distance_km']
```

```python
# Extração de ano e mês da data de início da locação
# Para a extração de informações de data temos
# dt.day
# dt.month
# dt.year
# dt.weekday
df_consolidado['dia'] = df_consolidado['start_date'].dt.month
df_consolidado['dia'] = df_consolidado['start_date'].dt.year
df_consolidado['dia'] = df_consolidado['start_date'].dt.weekday
```
